```python
# HDR Image들의 평균을 계산해서 Silhouette Image 만드는 프로그램
import matplotlib as plt
import rawpy
import PIL
import os
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci

def getfilelist(fname):
    path = "./"+fname+"/"
    file_list = os.listdir(path)
    ret = []
    for i in range(len(file_list)):
        Num = len(file_list[i])
        if Num > 4 and (file_list[i][-4:Num] == ".jpg" or file_list[i][-4:Num] == ".png"):
            ret.append([path,file_list[i]])
    return ret

def OpenImage(image):
    img = PIL.Image.open(image[0]+image[1])
    RGB = np.array(img)
    return (img, RGB)

def CloseImage(handle):
    handle[0].close()

os.makedirs("Output", exist_ok=True) # Result Images
Image_list = getfilelist("Input")
Cnt = 0
N = len(Image_list)
Silhouette =[]
Image_handle = []
print("HDR Images : {:d} Images are Loaded.".format(N))
for i, v in enumerate(Image_list):
    Image_handle.append(OpenImage(v))
    if Cnt == 0:
        Silhouette = Image_handle[i][1].astype('int32')
    else:
        Silhouette = np.add(Silhouette,Image_handle[i][1],dtype=np.int32)
    Cnt += 1
    print("{} processed.".format(v[1]))
    CloseImage(Image_handle[i])

Silhouette //= Cnt
Silhouette = np.clip(Silhouette, 0, 255).astype('uint8')
PIL.Image.fromarray(Silhouette).save("./Output/Silhouette.png")
print("Silhouette Mask Image ({}) is Processed.".format("Silhouette.png"))
```

```python
# monitor_pixels_h#w#.npy 생성 코드 (단위 : m)
import numpy as np

monitor = np.load('./monitor_pixels_h12w6_2.npy')
print(monitor.shape)
print((monitor*1000).astype('int32'))

# 카메라 위치 = (0,0,0)
# start (pattern000.png 기준) = [7*(X-4),13+7*Y,0] (16x9), [13*(X-2),15.5+13*Y,0] (9x5), [21.5*(X-1),12.75+21.5*Y,0] (6x3) [mm]
# [11*(X-2.5),7.5+11*Y,0] (12x6), [8*(X-3.5),6+8*Y,0] (16x8) [mm]
# superpixel(정사각형) 한 변의 길이 = 7mm (16x9), 13mm (9x5), 21.5mm (6x3), 11mm (12x6), 8mm (16x8), z = 0 고정(평면)

width = 6 # light_C
height = 12 # light_R
monitor =[[[0.0]*3 for _ in range(width)] for _ in range(height)]
for i in range(height):
    for j in range(width):
        monitor[i][j][0] = 0.011*(j-2.5) # X
        monitor[i][j][1] = 0.0075 + 0.011*(i) # Y
monitor = np.array(monitor).astype('float32')

np.save(f'./monitor_pixels_h{height}w{width}.npy', monitor)
print("monitor_pixels_h"+str(height)+"w"+str(width)+".npy is saved. {}\n".format(monitor.shape))
print(monitor)

```

```python
import matplotlib as plt
import rawpy
import PIL
import os
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
import cv2
from PIL import Image

def getfilelist(fname):
    path = "./"+fname+"/"
    file_list = os.listdir(path)
    ret = []
    for i in range(len(file_list)):
        Num = len(file_list[i])
        if Num > 4 and (file_list[i][-4:Num] == ".jpg" or file_list[i][-4:Num] == ".png"):
            ret.append([path,file_list[i]])
    return ret

def OpenImage(image):
    img = cv2.imread(image[0]+image[1],cv2.IMREAD_COLOR)
    return img

def CloseImage(handle):
    handle[0].close()

os.makedirs("Output", exist_ok=True) # Result Images
Image_list = getfilelist("20240820_155625_out")
Cnt = 0
N = len(Image_list)
print("HDR Images : {:d} Images are Loaded.".format(N))
for i, v in enumerate(Image_list):
    Image_handle = OpenImage(v)
    dstG = cv2.cvtColor(cv2.GaussianBlur(Image_handle,(0,0),1), cv2.COLOR_BGR2RGB)
    dstG *= 1
    Image.fromarray(dstG).save('./Output/'+v[1],'PNG')
    print("{} processed.".format(v[1]))
    Cnt += 1
print(f"Total {Cnt} Images are Processed.")
```
